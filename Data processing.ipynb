```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("best_Commpany_details_USA.csv")
display(df)
```

Replace $ and comma from columns

```python
columns_to_modify = ["Revenue", "Profit", "Assets", "employees"]

for col in columns_to_modify:
    df[col] = df[col].str.replace('$','', regex = False)
    df[col] = df[col].str.replace(',','', regex = False)
    df[col] = pd.to_numeric(df[col])

display(df)
```

Check for Common placeholder values in each columns of dataframe

```python
def check_placeholders(df, placeholders):
    for col in df.columns:
        placeholder_mask = df[col].isin(placeholders)
        if placeholder_mask.any():
            print(f"The {col} contains place holder values:")
            print(df[col][placeholder_mask])
        else:
            print(f"The {col} doesn't contains placeholders mask")

placeholders = ['-', 'N/A', '','NaN','Unknown']
check_placeholders(df, placeholders)
```

Remove % from columns as it denotes object (string)

```python
df["Revenue in percentage"]
```

```python
columns_to_rem_per = ["Revenue in percentage", "Profit in percentage"]
for col in columns_to_rem_per:
    df[col] = df[col].str.replace("%","",regex = False)
    df[col] = pd.to_numeric(df[col]) / 100
```

Handling NaN values

```python
columns_with_NaN = ['Revenue in percentage', 'Profit in percentage']

for col in columns_with_NaN:
    df[col] = df[col].ffill()
```
